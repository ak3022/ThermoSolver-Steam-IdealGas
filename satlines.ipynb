```python
import numpy as np
import matplotlib.pyplot as plt
import CoolProp.CoolProp as CP
from bokeh.models import Arrow, NormalHead
from bokeh.palettes import Category10
from bokeh.plotting import figure, show
from bokeh.plotting import row
#plotting the saturation lines for p-v diagram
Prange=np.linspace(611.655,2.2064e+07) #maybe the graph will look better depending on the actual process
densrange_x0=[]
densrange_x1=[]
vrange_x0=[]
vrange_x1=[]
for i in Prange:
    densrange_x0+=[CP.PropsSI('D','P', i ,'Q',0,"water")]
    densrange_x1+=[CP.PropsSI('D','P', i ,'Q',1,"water")]
for j in densrange_x0: #making v ranges
    vrange_x0+=[1/j]
for k in densrange_x1:
    vrange_x1+=[1/k]
#shortened vrangex1 and prange for v<0.5:
Prangeshortened=[]
vrangex1shortened=[]
for i in range(len(vrange_x1)):
    if vrange_x1[i]<0.5:
        vrangex1shortened+=[vrange_x1[i]]
        Prangeshortened+=[Prange[i]]
plotpv = figure(title="P-v plot", x_axis_label="v (m^3/kg)", y_axis_label="P (MPa)", width=550, height=300)
plotpv.line(vrange_x0,Prange, color="black", line_width=2)
plotpv.line(vrangex1shortened,Prangeshortened, color="black", line_width=2)

#plotting saturation lines for T-s diagram (note to self: coolprop gives temperature in kelvin)
Trange=np.linspace(273.06,647.096) 
s_range_x0=[]
s_range_x1=[]
for i in Trange:
    s_range_x0+=[CP.PropsSI('S','T', i ,'Q',0,"water")]
    s_range_x1+=[CP.PropsSI('S','T', i ,'Q',1,"water")]
plotts = figure(title="T-s plot", x_axis_label="s (J/kgK)", y_axis_label="T (K)", width=550, height=300)
plotts.line(s_range_x0,Trange, color="black", line_width=2)
plotts.line(s_range_x1,Trange, color="black", line_width=2)

#plotting saturation lines for h-s diagram 
Trange=np.linspace(273.06,647.096)
s_range_x0=[]
s_range_x1=[]
h_range_x0=[]
h_range_x1=[]
for i in Trange:
    s_range_x0+=[CP.PropsSI('S', 'T', i ,'Q',0,"water")]
    s_range_x1+=[CP.PropsSI('S', 'T', i ,'Q',1,"water")]
    h_range_x0+=[CP.PropsSI('H', 'T', i ,'Q',0,"water")]
    h_range_x1+=[CP.PropsSI('H', 'T', i ,'Q',1,"water")]
plot_hs = figure(title="h-s plot", x_axis_label="s (J/kgK)", y_axis_label="h (J/kg)", width=550, height=300)
plot_hs.line(s_range_x0,h_range_x0, color="black", line_width=2)
plot_hs.line(s_range_x1,h_range_x1, color="black", line_width=2)

#show(row(plotpv,plotts,plot_hs))
```
